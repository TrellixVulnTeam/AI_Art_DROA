```
#wczytanie bibliotek
%tensorflow_version 1.x
!pip install -q gpt-2-simple
import gpt_2_simple as gpt2
from datetime import datetime
from google.colab import files
#import tensorflow as tf
import time

#jak błąd z GPU to wejdź w Edytuj->Ustawienia Notatnika -> GPU
!nvidia-smi
```

```
gpt2.download_gpt2(model_name="355M")
```

```
gpt2.mount_gdrive()
```

```
file_name = "model_v01_clean.txt"
gpt2.copy_file_from_gdrive(file_name)
```

```
#tf.reset_default_graph()
sess = gpt2.start_tf_sess()
start = time.time()
gpt2.finetune(sess,
              dataset=file_name,
              model_name='355M',
              steps=15000,
              restore_from='fresh',
              run_name='mm_model_355M_v01',
              print_every=10,
              sample_every=2000,
              save_every=1000
              )
 
print(time.time() - start)
```

```
gpt2.copy_checkpoint_to_gdrive(run_name='mm_model_355M_v01')
```

```
gpt2.generate(sess,
              run_name='mm_model_355M_v01',
              length=150,
              temperature=0.7,
              prefix="Mój wszechmocny Mirosławie",
              nsamples=5,
              batch_size=5
              )
```
